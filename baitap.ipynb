```python
#Bai tap 1. 
#Biu  thanh nhm
#cho d liu sau:
labels=["Q1","Q2",'Q3','Q4']
A=[500,700,800,600]
B=[450,350,650,750]
C=[500,250,850,600]
#V biu  thanh nhm th hin doanh thu ca A,B v C trong 4 qu
import numpy as np
import matplotlib.pyplot as plt

# D liu
labels = ["Q1", "Q2", "Q3", "Q4"]
A = [500, 700, 800, 600]
B = [450, 350, 650, 750]
C = [500, 250, 850, 600]

# Xc nh v tr cc nhm
x = np.arange(len(labels))  # Ch mc ca cc qu
width = 0.25  #  rng ca mi thanh

# To biu 
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x - width, A, width, label='Doanh thu A', color='b')
ax.bar(x, B, width, label='Doanh thu B', color='g')
ax.bar(x + width, C, width, label='Doanh thu C', color='r')

# Gn nhn
ax.set_xlabel('Qu')
ax.set_ylabel('Doanh thu')
ax.set_title('Doanh thu theo qu ca A, B v C')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Hin th biu 
plt.show()
#Bai tap 2
#Biu  trn
Cty=["A",'B','C','D']
thiphan=[30,25,25,20]
color=['#ff9999','#66b3ff','#99ff99','#ffcc99']
#To biu  trn
# V biu  trn
plt.figure(figsize=(6, 6))
plt.pie(thiphan, labels=Cty, autopct='%1.1f%%', colors=color, startangle=140, wedgeprops={'edgecolor': 'black'})

# Tiu 
plt.title("T l th phn ca cc cng ty")

# Hin th biu 
plt.show()
#Bai tap 3.
#Biu  phn tn
x=np.random.rand(100)
y=np.random.rand(100)
sizes=np.random.rand(100)*300
colors=np.random.rand(100)
#To biu  phn tn ca 2 bin ngu nhin
plt.figure(figsize=(7, 5))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis', edgecolors='black')

# Gn nhn
plt.xlabel("X - Gi tr ngu nhin")
plt.ylabel("Y - Gi tr ngu nhin")
plt.title("Biu  phn tn vi kch thc v mu sc ngu nhin")

# Thm thanh mu
plt.colorbar(label="Mu sc")

# Hin th biu 
plt.show()


```
