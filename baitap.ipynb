```python
#Bai tap 1. 
#Biểu đồ thanh nhóm
#cho dữ liệu sau:
labels=["Q1","Q2",'Q3','Q4']
A=[500,700,800,600]
B=[450,350,650,750]
C=[500,250,850,600]
#Vẽ biểu đồ thanh nhóm thể hiện doanh thu của A,B và C trong 4 quý
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu
labels = ["Q1", "Q2", "Q3", "Q4"]
A = [500, 700, 800, 600]
B = [450, 350, 650, 750]
C = [500, 250, 850, 600]

# Xác định vị trí các nhóm
x = np.arange(len(labels))  # Chỉ mục của các quý
width = 0.25  # Độ rộng của mỗi thanh

# Tạo biểu đồ
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x - width, A, width, label='Doanh thu A', color='b')
ax.bar(x, B, width, label='Doanh thu B', color='g')
ax.bar(x + width, C, width, label='Doanh thu C', color='r')

# Gán nhãn
ax.set_xlabel('Quý')
ax.set_ylabel('Doanh thu')
ax.set_title('Doanh thu theo quý của A, B và C')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Hiển thị biểu đồ
plt.show()
#Bai tap 2
#Biểu đồ tròn
Cty=["A",'B','C','D']
thiphan=[30,25,25,20]
color=['#ff9999','#66b3ff','#99ff99','#ffcc99']
#Tạo biểu đồ tròn
# Vẽ biểu đồ tròn
plt.figure(figsize=(6, 6))
plt.pie(thiphan, labels=Cty, autopct='%1.1f%%', colors=color, startangle=140, wedgeprops={'edgecolor': 'black'})

# Tiêu đề
plt.title("Tỷ lệ thị phần của các công ty")

# Hiển thị biểu đồ
plt.show()
#Bai tap 3.
#Biểu đồ phân tán
x=np.random.rand(100)
y=np.random.rand(100)
sizes=np.random.rand(100)*300
colors=np.random.rand(100)
#Tạo biểu đồ phân tán của 2 biến ngẫu nhiên
plt.figure(figsize=(7, 5))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis', edgecolors='black')

# Gán nhãn
plt.xlabel("X - Giá trị ngẫu nhiên")
plt.ylabel("Y - Giá trị ngẫu nhiên")
plt.title("Biểu đồ phân tán với kích thước và màu sắc ngẫu nhiên")

# Thêm thanh màu
plt.colorbar(label="Màu sắc")

# Hiển thị biểu đồ
plt.show()


```
